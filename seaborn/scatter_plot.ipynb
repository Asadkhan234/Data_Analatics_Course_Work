```python
# loading libraries:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data set:
data = sns.load_dataset('tips')
data.head(1)
```

```python
# Plot simple scatter plot:
d=data.groupby('total_bill')['tip'].sum().reset_index()
#print(d.max())
#print(d.min())
plt.subplot(2,2,1)
sns.scatterplot(data=d,x='total_bill',y='tip')
plt.title('Plot 1')

plt.subplot(2,2,2)
sns.scatterplot(data,x='total_bill',y='tip',hue='sex',size='size')
plt.title('Plot 2')
plt.legend(loc='best',ncol=2,bbox_to_anchor=(1,1))


plt.subplot(2,2,3)
sns.scatterplot(data,x='total_bill',y='tip',hue='smoker')
plt.title('Plot 3')

plt.subplot(2,2,4)
sns.scatterplot(data,x='total_bill',y='tip',hue='day')
plt.title('Plot 4')
plt.legend(loc='best',ncol=1,bbox_to_anchor=(0.4,0.4))

plt.tight_layout()
plt.show()
```
