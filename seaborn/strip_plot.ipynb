```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# used where one data is string and one is numeric

```python
data = sns.load_dataset('tips')
data.head(3)
```

```python
sns.stripplot(data,x='sex',y='size',hue='smoker',dodge=True,jitter=0.2)
plt.show()
```

```python
sns.stripplot(data,x='time',y='size',hue='sex',jitter=0.3,dodge=True)
plt.show()
```
