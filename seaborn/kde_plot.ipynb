```python
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data.head(1)
```

```python
sns.kdeplot(data,x='total_bill',hue='sex',multiple='fill',common_grid=True,)
plt.show()
```

```python
sns.kdeplot(data,x='tip',hue='size',multiple='stack',palette='viridis')
plt.show()
```
