```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('titanic')
data.head(3)
```

```python
sns.swarmplot(data,x='alive',y = 'age',hue='sex',size=4,edgecolor='black')
plt.show()
```

```python
sns.swarmplot(data,x='class',y='survived',hue='sex')
plt.show()
```

```python
sns.swarmplot(data,x='who',y='alive',hue='sex')
plt.show()
```
