```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data.head(3)
```

```python
a  = sns.FacetGrid(data,col='sex',hue='size',palette='viridis')
a.map(sns.barplot,'time','tip')
plt.show()
```

```python
b = sns.FacetGrid(data,col = 'day',hue='smoker')
b.map(sns.barplot,'sex','total_bill')
plt.show()
```
