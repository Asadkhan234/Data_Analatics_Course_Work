```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data.head(2)
```

```python
sns.countplot(data=data,x='smoker',hue='sex')
plt.show()
```

```python
sns.countplot(data=data,x='day',hue='tip')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data,x='time',hue='tip')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data,x='time',hue='sex')
plt.show()
```

```python
data = pd.read_excel('ESD (3).xlsx')
data.head(2)
```

```python
sns.countplot(data=data,x='Gender',hue='Country',saturation=2,width=1,gap=0.3,order=['Male','Female'])
plt.show()
```

```python
sns.countplot(data,x='Ethnicity',hue='Age',saturation=2,gap=0.2,color='red')
plt.show()
```

```python
d = data.groupby('Full Name')['Hire Date'].mean().reset_index().head(10)
plt.figure(figsize=(14,5))
sns.countplot(data=d,x='Full Name',hue='Hire Date')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True,color='green',linestyle='--',linewidth=1.5)
plt.title('Top 10 Employers')
plt.show()
```

```python
d = data.groupby('Job Title')['Hire Date'].mean().reset_index().head(10)
plt.figure(figsize=(25,10))
sns.countplot(data=d,x='Job Title',hue='Hire Date',width=0.8)
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
plt.grid(True,color='green',linestyle='--',linewidth=1.5)
plt.title('Top 10 Employers')
plt.show()
```
