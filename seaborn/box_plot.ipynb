```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data.head(2)
```

```python
sns.boxplot(data,x='sex',y='tip',hue='day',orient='x',dodge=True,linecolor='black',linewidth=2, fliersize=10)
plt.title('Sex vs Tips')
plt.xlabel('Sex')
plt.ylabel('Tips')
plt.show()
```

```python
sns.boxplot(data,x='day',y='total_bill',hue='time',orient='x',linecolor='black',linewidth=2, fliersize=10)
plt.title('Total vs Days')
plt.xlabel('Total')
plt.ylabel('Bills')
plt.show()
```
