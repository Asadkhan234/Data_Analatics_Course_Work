```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data.head(1)
```

```python
sns.histplot(data, x='day',hue='sex',kde=True)
plt.show()
```

```python
data = sns.load_dataset('titanic')
data.head(1)
```

```python
sns.histplot(data=data,x='age',hue='who',kde=True,bins=30,discrete=True)
plt.grid(True)
plt.show()
```

```python
sns.histplot(data,x='sex',hue='alive',kde=True)
plt.grid(True)
plt.show()
```

```python
sns.histplot(data,x='who',y='embark_town',kde=True,label='embark_town')
plt.legend()
plt.grid(True)
plt.show()
```
