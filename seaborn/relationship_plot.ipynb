```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data.head(3)
```

```python
sns.relplot(data,x='tip',y='total_bill',hue='sex',col='smoker',size='day')
plt.show()
```

```python
sns.relplot(data,x='tip',y='size',hue='sex',col='day',size='time')
plt.show()
```
