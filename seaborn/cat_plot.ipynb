```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Cat Plot stand for categorical data 

```python
data = sns.load_dataset('tips')
data.head(3)
```

```python
sns.catplot(data,x='sex',y='tip',hue='time')
plt.show()
```

```python
sns.catplot(data,x='sex',y='tip',hue='time',kind='violin')
plt.show()
```

```python
sns.catplot(data,x='sex',y='tip',hue='time',kind='bar',errorbar=('ci',0))
plt.show()
```

```python
sns.catplot(data,x='sex',hue='time',kind='count')
plt.show()
```

```python
sns.catplot(data,x='sex',y='tip',hue='time',kind='point',)
plt.show()
```
