```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data.head(3)
```

```python
# REMEMBER IT IS NOT WORKING ON SERIES DATA , YOU SHOULD CONVERT IT INTO .to_frame()
```

```python
d = data.groupby('day', observed=True).agg({'tip':'mean'})
sns.heatmap(d)
plt.show()
```

```python
c = data.groupby('sex',observed=True)['tip'].sum().to_frame()
sns.heatmap(c)
plt.show()
```

```python
data = pd.read_excel('ESD (3).xlsx')
data.head(2)
```

```python
salary = data.groupby('Job Title').agg({'Annual Salary':'mean','Bonus %':'mean'})
# Make it numeric matrix for heatmap
sns.heatmap(salary, annot=True, cmap='Blues', fmt=".1f")

plt.title("Average Salary & Bonus % by Job Title")
plt.ylabel("Job Title")
plt.show()
```

```python
Age = data.groupby(['Gender','Country']).agg({'Age':'mean'})
sns.heatmap(data=Age,annot=True,cmap='Blues',fmt='.1f')
```
