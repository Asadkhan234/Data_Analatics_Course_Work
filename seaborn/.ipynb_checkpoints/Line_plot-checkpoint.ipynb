```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame({
    'Days' : [1,2,3,4,5,6,7],
    'Nop' : [40,30,60,70,100,80,90]
})
data
```

```python
sns.lineplot(data=data,x=data['Days'],y=data['Nop'])
plt.grid(True)
plt.show()
```

```python
mobiles = pd.DataFrame({
    'Brands':['Iphone','Samsung','Oppo','Vevo','Infinix'],
    'Prices':[100000,80000,70000,50000,30000]
})
mobiles
```

```python
sns.lineplot(data=mobiles,x= mobiles['Brands'],y= mobiles['Prices'],label='Brand-Prices')
plt.legend(loc='best')
plt.grid(color='red',linestyle='--',drawstyle='steps-mid')
plt.show()
```

```python
data = pd.read_excel('ESD (3).xlsx')
data.info()
```

```python
sns.lineplot(data=data , x ='Annual Salary',y='Country')
plt.show()
```

```python
data.head(1)
```

```python
sns.lineplot(data = data , x = 'Business Unit',y='Age',hue='Country',markers='--',)
plt.show()
```

```python
sns.lineplot(data=data , x = 'Hire Date', y = 'Exit Date',hue='Gender')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
d = data.groupby(['Department','Gender'])['Annual Salary'].count().reset_index()
sns.lineplot(data=d ,x = 'Department', y = 'Annual Salary',marker='o',hue='Gender')
plt.grid(color = 'green',linestyle='--',linewidth=1.3)
plt.show()
```

```python
color = ['red','green','yellow']
sns.lineplot(data = data , x = data['Ethnicity'],y  = 'Country',hue = 'Gender',palette=color,marker='o')
plt.grid(color = 'gray',linestyle = '--',linewidth = 2)
plt.show()
```
