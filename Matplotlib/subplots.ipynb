```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5,6]
y1 = [20,40,70,30,80,60]

x2 = [1,2,3,4,5,6]
y2 = [30,50,70,90,10,40]

x3 = [1,2,3,4,5,6]
y3 = [50,30,20,40,60,80]

# 1st subplot (row=2, col=2, position=1)
plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'b')
plt.title("Plot 1")

# 2nd subplot (row=2, col=2, position=2)
plt.subplot(2, 2, 2)
plt.plot(x2, y2, 'r')
plt.title("Plot 2")

# 3rd subplot (row=2, col=2, position=3)
plt.subplot(2, 2, 3)
plt.plot(x3, y3, 'g')
plt.title("Plot 3")

plt.suptitle('Three Plots Of Ages')
plt.tight_layout()


# 4th subplot (row=2, col=2, position=3)
plt.subplot(2,2,4)
plt.plot(x1,y3)
plt.title('plot 4')
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_csv('student_scores.csv')
data.head(1)
```

```python
import matplotlib.pyplot as plt

# Subplot 1 (top-left)
plt.subplot(2, 2, 1)  # 2 rows, 2 cols, position 1
plt.bar(data['MathScore'], data['ReadingScore'], label='Math Vs Reading')
plt.title('Math Vs Reading')
plt.legend(loc='upper left')   # avoid "best"

# Subplot 2 (top-right)
plt.subplot(2, 2, 2)  # 2 rows, 2 cols, position 2
plt.scatter(data['MathScore'], data['WritingScore'], label='Math Vs Writing')
plt.title('Math Vs Writing')
plt.legend(loc='upper left')   # avoid "best"

plt.show()
```
