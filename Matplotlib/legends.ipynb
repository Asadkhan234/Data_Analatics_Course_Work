```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,50,60,70]
y = [30,40,60,80,100,120]
y1 = [20,10,40,50,70,90]
y2 = [10,30,40,50,90,100]
y3 = [100,50,60,30,40,10]

plt.plot(x,y,label = 'male')
plt.plot(x,y1,label = 'Female')
plt.plot(x,y2,label='Kids')
plt.plot(x,y3,label='blanks')
plt.legend(bbox_to_anchor=(0.9,1.1),ncols=4,labelspacing =5)
#plt.legend(loc=9)
#plt.legend(['a','b','c'],ncols = 3) # loc is used for location of legend.
plt.show()
```
